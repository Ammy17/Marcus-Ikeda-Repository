```python
query = """ SELECT distinct substring("Order Date",1,4) AS "OrderYear"
            FROM df
            ORDER by OrderYear

        """
a = sqldf(query)
a = a.values.tolist()

years = []
for item in a :
    years.append(item[0])
years
```

```python

query = """     
            SELECT substring("Order Date",1,4) as OrderYear, COUNT(*) as Count
            FROM df 
            WHERE State = 'Nevada'
            GROUP BY substring("Order Date",1,4)

        """
b = sqldf(query)
b
```

```python
query = """ SELECT DISTINCT state
            FROM df
        """
a = sqldf(query)
a = a.values.tolist()

states = []
for item in a :
    states.append(item[0])
states
```

```python
query = """     
                SELECT State, substring("Order Date",1,4) as 'Order Year',  count('Order Date') as 'Number of Orders'
                FROM df
  
                GROUP BY substring("Order Date",1,4), State
                HAVING count('Order Date') < 5
                ;
        """
b = sqldf(query)
b
```

```python
df['OrderYear'] = df['Order Date'].dt.strftime('%Y')
```

```python
df2 = df.filter(['OrderYear','State','Sub-Category'], axis=1)
```

```python

pivot = df2.groupby(['State','OrderYear',]
                 ,as_index = False
                ).count().pivot('State'
                    ,'OrderYear').fillna(0)
display(pivot)

```

```python

result = pivot.loc[((pivot[('Sub-Category', '2015')]) < 5) |
                  ((pivot[('Sub-Category', '2016')]) < 5) | 
                  ((pivot[('Sub-Category', '2017')]) < 5) | 
                  ((pivot[('Sub-Category', '2018')]) < 5)]
result
```
